```python
import numpy as np
from scipy import special as ss
import matplotlib.pyplot as plt
```

Logit function

```python
x = np.linspace(0, 1, 501)
y = ss.logit(x)
plt.plot(x, y)
plt.grid()
plt.ylim(-6, 6)
plt.xlabel('x')
plt.title('logit(x)')
plt.show()
```

L

```python
fig, ax = plt.subplots(figsize=(10,9))
x = np.linspace(-1, 1, 501)
for i, j in enumerate((0,1,2,5,10,100)):
    plt.subplot(3, 2, i+1)
    y = ss.eval_legendre(j,x)
    plt.plot(x, y, label = "legendre(x,"+str(j)+")")
    plt.grid()
    # plt.ylim(-6, 6)
    plt.xlabel('x')
    plt.legend()
```
